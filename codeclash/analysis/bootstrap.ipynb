```python
def analysis(
    true_elos: list[float],
    *,
    n_rounds: int = 10,
    k: float = 32,
    update_strategy: Literal["per_round", "per_tournament"] = "per_round",
    game_reps: int = 1_000,
    tournament_reps: int = 10,
    elo_start: float = 1_500,
):
    players = [Player(name=f"P{i}", true_elo=elo, measured_elo=elo_start) for i, elo in enumerate(true_elos)]
    game = Game("g1", draw_probability=0.0, repetitions=game_reps)
    tournament = Tournament(game, n_rounds=n_rounds, k=k, update_strategy=update_strategy)
    analysis = TwoPlayerBasedLeaderboard([tournament] * tournament_reps)
    analysis.run(players)
    print([p.measured_elo for p in players])
```

```python
analysis([1000, 2000])
analysis([1000, 2000], update_strategy="per_tournament", k=300)
analysis([1000, 2000], k=60)
analysis([1000, 2000], tournament_reps=20)
```

```python
true_elos = [1000, 1500, 1750, 2000]
analysis(true_elos)
analysis(true_elos, update_strategy="per_tournament", k=300)
analysis(true_elos, k=60)
analysis(true_elos, tournament_reps=20)
```
